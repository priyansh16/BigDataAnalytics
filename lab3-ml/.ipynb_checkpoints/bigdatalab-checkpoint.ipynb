```python
import numpy as np
import matplotlib.pyplot as plt

# Distance kernel
h1 = 200
x1 = np.arange(-1000, 1000, 1)
y1 = np.exp(-(x1/h1)**2)

plt.figure(figsize=(8, 6))
plt.plot(x1, y1, label='Distance kernel')
plt.xlabel('Possible distances')
plt.ylabel('Kernel')
plt.title('Distance Kernel')
plt.savefig("figs/DistanceKernel.png")
plt.show()
```

```python

# Time kernel
h2 = 2
x2 = np.arange(-12, 12, 0.3)
y2 = np.exp(-(x2/h2)**2)

plt.figure(figsize=(8, 6))
plt.plot(x2, y2, label='Time kernel')
plt.xlabel('Possible time differences')
plt.ylabel('Kernel')
plt.title('Time Kernel')
plt.savefig("figs/TimeKernel.png")
plt.show()
```

```python
# Date kernel
h3 = 12
x3 = np.arange(-180, 180, 1)
y3 = np.exp(-(x3/h3)**2)

plt.figure(figsize=(8, 6))
plt.plot(x3, y3, label='Date kernel')
plt.xlabel('Possible day differences')
plt.ylabel('Kernel')
plt.title('Date Kernel')
plt.savefig("figs/DateKernel.png")
plt.show()
```
